```python
import fitz

pdf_path = "document.pdf"
doc = fitz.open(pdf_path)
text = ""

for page_num in range(doc.page_count):
    page = doc[page_num]
    text += page.get_text()

doc.close()
```

```python
def extract_abstract(parsed_text):
    # Find the index of "Abstract"
    start_index = parsed_text.find("abstract")
    
    # Find the index of "Keywords"
    end_index = parsed_text.find("keywords")
    
    # Extract the abstract if both start and end indices are found
    if start_index != -1 and end_index != -1:
        abstract = parsed_text[start_index +8:end_index].strip()
        return abstract
    else:
        return 0

# Example usage:
# parsed_text = "Some text abstract This is keywords"
# abstract = extract_abstract(parsed_text)

# if abstract:
#     print("Abstract:\n", abstract)
# else:
#     print("Abstract not found.")
```
